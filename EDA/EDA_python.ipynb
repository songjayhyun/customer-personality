```python
# 결혼 여부와 소득같의 상관계수 파악
# 유의 수준 0.05  기준으로 기각 , 유의미하지 않음으로 판단

from scipy.stats import pearsonr

correlation_coefficient, p_value = pearsonr(dropna_df['소득'], pd.factorize(dropna_df['결혼여부'])[0])

# 결과 출력
print(f"상관계수: {correlation_coefficient}")
print(f"p_value: {p_value}")
```

```python
# 학력와 소득같의 상관계수 파악
# 0.05 미만이기때문에 유의미함
# 소득 결측치를 교육수준 기준으로 평균내 결측치 채우기


from scipy.stats import pearsonr

correlation_coefficient, p_value = pearsonr(dropna_df['소득'], pd.factorize(dropna_df['교육수준'])[0])

# 결과 출력
print(f"상관계수: {correlation_coefficient}")
print(f"p_value: {p_value}")
```

```python
if p_value<0.05:
  print('참')
```
