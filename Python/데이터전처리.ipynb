```python
#필요없는 컬럼 삭제


df.drop(['Z_CostContact','Z_Revenue'],axis=1,inplace=True)
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
# 데이터 생성 연도인 2021년 기준으로 나이(age) 컬럼 생성

df['age']=2021-df['Year_Birth']
```

```python
df.head()
```

```python
from scipy.stats import chi2_contingency


# 카이제곱 검정
crosstab = pd.crosstab(df['Marital_Status'], df['Income'])
chi2, p, _, _ = chi2_contingency(crosstab)

# 결과 출력
print("Chi-square value:", chi2)
print("p-value:", p)

# 결과 해석
alpha = 0.05
if p < alpha:
    print("결혼 여부와 수입 간에는 통계적으로 유의미한 관계가 있다. (귀무가설 기각)")
else:
    print("결혼 여부와 수입 간에는 통계적으로 유의미한 관계가 없다. (귀무가설 채택)")
```

```python
from scipy.stats import chi2_contingency
# 귀무가설 : 교육수준과 수입 간에 관계가 없다.

crosstab = pd.crosstab(df['Education'], df['Income'])
chi2, p, _, _ = chi2_contingency(crosstab)


# 0.05 넘으면 채택, 아니면 기각
if p<0.05:
	print('귀무가설 기각 : 교육수준과 수입간에 유의미한 관계있다.')
else:
	print('귀무가설 채택 : 교육수준과 수입간에 유의미한 관계없다.')
```
